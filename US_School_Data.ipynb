```python
df_NYC = pd.read_csv(files_NYC[1])
```

```python
df_CA = pd.read_csv(files_CA[1])
```

```python
# both CA and NYC files have missing data for years 2017 and 2018, so consider only data from 2008-2016
df_NYC = df_NYC[df_NYC.math_test_pct_prof_midpt.notna()]
df_CA = df_CA[df_CA.math_test_pct_prof_midpt.notna()]

df_NYC = df_NYC[df_NYC.read_test_pct_prof_midpt.notna()]
df_CA = df_CA[df_CA.read_test_pct_prof_midpt.notna()]
```

```python
mapping = {'Suppressed data': 0,'Missing/not reported':0}
df_NYC.read_test_pct_prof_midpt.replace({'Suppressed data': mapping})
df_CA.read_test_pct_prof_midpt.replace({'Suppressed data': mapping})


df_NYC.math_test_pct_prof_midpt.replace({'Suppressed data': mapping})
df_CA.math_test_pct_prof_midpt.replace({'Suppressed data': mapping})


df_NYC.guidance_counselors_sec_fte.replace({'Missing/not reported':mapping})
df_CA.guidance_counselors_sec_fte.replace({'Missing/not reported':mapping})


df_CA.teachers_secondary_fte.str.replace(r'[^0-9]+','')
df_CA.math_test_pct_prof_midpt.str.replace(r'[^0-9]+','')
df_CA.read_test_pct_prof_midpt.str.replace(r'[^0-9]+','')

df_NYC.teachers_secondary_fte.str.replace(r'[^0-9]+','')
df_NYC.math_test_pct_prof_midpt.str.replace(r'[^0-9]+','')
df_NYC.read_test_pct_prof_midpt.str.replace(r'[^0-9]+','')

```

```python
df_CA.math_test_pct_prof_midpt._get_numeric_data()
df_CA.read_test_pct_prof_midpt._get_numeric_data()
df_CA.teachers_secondary_fte._get_numeric_data()


df_NYC.math_test_pct_prof_midpt._get_numeric_data()
df_NYC.read_test_pct_prof_midpt._get_numeric_data()
df_NYC.teachers_secondary_fte._get_numeric_data()
```

```python
x = df_NYC.year.unique().tolist()
```

```python

y = df_NYC.groupby('year')['enrollment'].count()
```

```python
fig1 = plt.figure()


ax1 = fig1.add_subplot(1,1,1)
# multiple plots in the same subplot
ax1.plot(x,y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
#ax1.set_title("XY Plot")
plt.show()
```
